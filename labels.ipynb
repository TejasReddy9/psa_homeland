First, I want to read in the 'stage1_labels.csv' file into a Pandas dataframe (labels)

```python
labels = pd.read_csv('/Users/Yvonne/Desktop/TSA_Kaggle/stage1_labels.csv')
labels.head()
```

Because this dataset is huge (19,499 entries and about 3 TB of data) I want to focus first on training models for a single body zone (of 17). In the csv file Kaggle provided, the zone information is tacked onto the end of the unique ID for each entry in the 'Id' column. I want to split the 'Id' column 

```python
labels['Id_only'], labels['Zone'] = labels['Id'].str.split('_', 1).str
labels.head()
```

Now that 'Zone' is its own column, it's much easier to sort and index the labels dataframe by zone. I want to create a new dataframe (zone6) that includes only the row entries in zone 6. This is the subset of the dataset that I'll start working on to try my model. I'll pull an initial sample of entries from zone 6 to begin with.

```python
zone6 = labels[labels['Zone']=='Zone6']
zone6.head()
```
